### Stemming
It is the process of reducing the infected words to their word stem.

Ex. history, historical will be reduced to histori
finally, final, finalized will be reduced to fina
going, goes, gone -> go

Used in Sentiment Analysis, word prediction
helps us determine whether the word is positive or negative.

### Lemmatization
It also does reduction of the infected words into meaningful word stem.

Ex. history, historical -> history,
finally, final, finalized -> final
going, goes, gone -> go

#### Basic Difference:
Lemmatized word will always have some meaning, it takes some time, that are opposite for stemming.

Stemming used in sentiment classifier, gmail classifer(dont need to have meaningful token).

Lemmatization used in chatbots, QnA apps.

```python
# Stemming
from nltk.stem import PorterStemmer # library
from nltk.corpus import stopwords # Library
```

```python
print(para)
```

```python
sentences = nltk.sent_tokenize(para)
stemmer = PorterStemmer()
```

```python
# stopwords are the words that dont hold much value in a statement.
print(stopwords.words("english"))
```

```python
# Now we will stem all the sentences one by one.
for i in range(len(sentences)):
    wordList = nltk.word_tokenize(sentences[i])
    wordList = [stemmer.stem(word) for word in wordList if word not in set(stopwords.words("english")) ]
    sentences[i] = ' '.join(wordList)
#      = wordList
```

```python
# The words produced may not have any meaning.
print(sentences)
```

```python
# Lemmatization
from nltk.stem import WordNetLemmatizer
```

```python
sentences = nltk.sent_tokenize(para)
lemmatizer = WordNetLemmatizer()
```

```python
for i in range(len(sentences)):
    wordList = nltk.word_tokenize(sentences[i])
    wordList = [lemmatizer.lemmatize(word) for word in wordList if word not in set(stopwords.words("english"))]
    sentences[i] = " ".join(wordList)

print(sentences)
```

### Bag of Words

<br>Disadvantage: 
With the values are 1, 0, etc. the semantics are not clear from the vector.
For good and boy we have the frequencies as 1 and 1 only. That does not tell us which word is more important or which word will precede the other word, etc.

TF-IDF, Word2Vec, GloVe beat it.

```python
import re
```

```python
sentences = nltk.sent_tokenize(para)
corpus = list()
```

```python
# Creating BOW with stemming
for i in range(len(sentences)):
    review = re.sub("[^a-zA-Z]", ' ', sentences[i])
    review = review.lower()
    review = review.split()
    review = [stemmer.stem(word) for word in review if word not in set(stopwords.words("english"))]
    review = ' '.join(review)
    corpus.append(review)

print(corpus)
```

### TF-IDF Term Frequency - Inverse Document Frequency
Similar to Bag of Words, we create vectors for each sentence and then instead of directly using frequency in the vectors, we instead use term frequency * inverse document frequency product as the value in the vector.

$ Term Frequency = \frac{No. of occurrence of word}{total no. of words in sentence} $<br>
$ Inverse Document Frequency = log_{10}(\frac{No. of sentences}{No. of sentences containing words}) $

Taking the same example:

| Sentences | Lemmatized |
| --- | --- |
| Sent 1 | good boy |
| Sent 2 | good girl |
| sent 3 | boy girl good |

##### Calculating the Term Frequencies 

|  | Sent 1 | Sent 2 |  Sent 3 |
| --- | --- | --- | --- |
| good | 1/2 | 1/2 | 1/3 |
| boy | 1/2 | 0 | 1/3 |
| girl | 0 | 1/2 | 1/3 |

##### Calculating the Inverse Document Frequencies

| Words | IDF |
| --- | --- |
| good | log(3/3) = 0 |
| boy | log(3/2) |
| girl | log(3/2) |

##### Final Vectors for all the sentences

| Sentences | good | boy | girl |
| --- | --- | --- | --- |
| sent 1 | (1/2)* 0 | (1/2) * log(3/2) | 0 |
| sent 2 | 0 | 0 | (1/2)* log(3/2) |
| sent 3 | 0 | (1/3)* log(3/2) | (1/3)* log(3/2) |

Hence, "good" should be given less importance, more importance to "boy" and "girl"

#### Problems with Bag of Words and TF IDF
In both BOW and TFIDF, semantic information is not stored.
TFIDF gives more importance to uncommon words.

There is definitely the chance of over fitting in case of implementation via Neural Networks.

### Word Embeddings
(Feature Representation)

-> *It is a technique to map words with their feature vectorr of real numbers. Word Embeddings are capable of capturing meaning of words, semantic and syntactic similarity*

-> With disadvantages of BOW and TFIDF, Word Embeddings is an improved technique where the semantic and syntactic information of a document(sentence) can be captured.

-> Implemented using Word2Vec and GloVe.

-> Also implemented using Embedding layers in Keras.

#### Feature vector.
Below is the representation of feature vectors for different words that have some relation in between them.

*A feature vector is an array where each index stores the magnitude of feature that the word has. It has its dimension(size) that means how many features are defining a single entity(word) and together all features represent the word.*

*A feature vector is a vector that stores the features for a particular observation in a specific order.*

For instance, boy and king have similar values of Gender but different values for royal feature.

| features | word1 | word2 | word3 | word4 |
| --- | --- | --- | --- | -- |
|  | Boy | Girl | King | Queen |
| Gender | -1 | 1 | -0.92 | .93 |
| Royal | 0.01 | 0.02 | 0.95 | 0.96 |
| Age | 0.03 | 0.02 | 0.7 | 0.6 |
| Food | 0.8 | 0.9 | 1.0 | 1.0 |
| . |
| . |
| . |

##### One hot representation: 
"this place is good" -> [300, 1000, 3000, 7000]
suppose the corpus(the set of all the words used in the dataset) has 10,000 words, and the indices of this, place, is, good respectively are 300, 1000, 3000, 7000.
It will be like: [ [0,0,0,...1(300th),...0,0,0,0], ... ].

The vector representation in this manner is very sparse (1's are seperated by a large distance in a vector) and it increases the complexity of computation because of large vector size(high dimension).

Not much semantic information can be saved in one hot vector format.

Hence, we use a set of features that collectively make a vector of real numbers and that represents a word.

It represents a word in 3D space. 

Similar words lie near each other in 3D space.

With the help of cosine similarity and cosine distance, similar words can be predicted as well.

Boy -> Girl

King -> ? 

We know the answer is queen, but it will be calculated based on (Boy)(->) - (Girl)(->) ~ (King)(->) vector.


source:(https://www.researchgate.net/publication/332892222_TMSA_A_Mutual_Learning_Model_for_Topic_Discovery_and_Word_Embedding/figures?lo=1)

But it is difficult to say exactly what features has the model captured from the word corpus. Only the machine knows.

Important Parameters: vocabulary size, one hot vector array, dimensions of embedding layer.

```python
!pip install tensorflow
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sents = ["the glass of milk", "the glass of juice", "the cup of tea", "I am a good boy", "I am a good developer"]
vocab_size = 10000
```

```python
oneHotReps = [ one_hot(words, vocab_size) for words in sents]
print(oneHotReps)
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
dim = 15

sent_length = 8

list_to_embedd = pad_sequences(oneHotReps, padding="pre", maxlen = sent_length)
print(list_to_embedd)
```

```python
model = Sequential()
model.add(Embedding(vocab_size, dim, input_length=sent_length))
model.compile("adam", "mse")
```

```python
model.summary()
# find how to count parameters in embedding.
```

```python
print(model.predict(list_to_embedd))
```

```python
text = re.sub(r"\[[0-9]*\]", ' ', para)
text = re.sub(r"\s+", " ", text)
text = text.lower()
text = re.sub(r"\d", " ", text)
text = re.sub(r"\s+", " ", text)

print(text)
```

```python
sentences = nltk.sent_tokenize(text)
sentences = [nltk.word_tokenize(sent) for sent in sentences]

for i in range(len(sentences)):
    sentences[i] = [lemmatizer.lemmatize(word) for word in sentences[i] if word not in stopwords.words("english")]

print(sentences)
```

```python
model = Word2Vec(sentences, min_count=1)
words = model.wv.key_to_index
print(words)
vector = model.wv['space']
print(vector, len(vector))
print(model.wv.key_to_index['space'])
similar = model.wv.most_similar("space")
print(similar)
```
